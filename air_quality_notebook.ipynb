```python
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("air_quality_preprocessed_dataset.csv")

# boxplot
sns.boxplot(data=df, x='city', y='aqi')
plt.xticks(rotation=45)
plt.title('AQI Distribution by City')
plt.show()

# heatmap
numeric_cols = df.select_dtypes(include=['number'])
sns.heatmap(numeric_cols.corr(), annot=True, cmap='RdYlBu')
plt.title('Air Quality Correlations')
plt.show()

# distribution of AQI
sns.histplot(data=df, x='aqi', kde=True)
plt.title('AQI Distribution')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("air_quality_preprocessed_dataset.csv", encoding="utf-8-sig")

# create the matplotlib plot
plt.figure()

city_avg = df.groupby('city')['aqi'].mean().sort_values(ascending=False)

bars = plt.bar(city_avg.index, city_avg.values)

plt.title("Average Air Quality Index by City")
plt.xlabel("City")
plt.ylabel("Average AQI")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# plotting histogram with matplotlib
df.plot.hist()
plt.show()

# plotting scatter with matplotlib
df.plot.scatter('city', 'aqi')
plt.xticks(rotation=45, ha='right')
plt.show()

# plotting boxplot with matplotlib
df.boxplot('aqi')
# print(df.columns.tolist())
plt.show()
```
